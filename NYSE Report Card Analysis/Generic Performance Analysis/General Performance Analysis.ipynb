#### Overall Proficiency Rate Across All Subjects

```python
import pandas as pd
```

```python
# Load data
data_ela = pd.read_excel(r'E:\Data Analytics\NYSE Report Card\Annual_EM_ELA.xlsx')
```

```python
data_math = pd.read_excel(r'E:\Data Analytics\NYSE Report Card\Annual_EM_MATH.xlsx')
```

```python
data_science = pd.read_excel(r'E:\Data Analytics\NYSE Report Card\Annual_EM_SCIENCE.xlsx')
```

```python
data_ela.head()
```

```python
data_math.head()
```

```python
data_science.head()
```

```python
# Combine data
all_subjects = pd.concat([data_ela, data_math, data_science])
```

```python
# Overall proficiency rate
all_subjects['PER_PROF'] = pd.to_numeric(all_subjects['PER_PROF'], errors = 'coerce')
```

```python
overall_proficiency = all_subjects['PER_PROF'].mean()
```

```python
print(f"Overall Proficiency Rate: {overall_proficiency:.2f}%")
```

#### Highest Proficiency Rate Across Demographic Subgroups


```python
# Group by subgroup and calculate mean proficiency
subgroup_proficiency = all_subjects.groupby("SUBGROUP_NAME")["PER_PROF"].mean()
subgroup_proficiency
```

```python
top_subgroup = subgroup_proficiency.idxmax()
top_subgroup
```

```python
top_subgroup_rate = subgroup_proficiency.max()
top_subgroup_rate
```

```python
print(f"Highest Proficiency Rate: {top_subgroup} with {top_subgroup_rate: .2f}%")
```

#### Performance Distribution Across Levels

```python
# Compute performance levels distribution
performance_levels = all_subjects[["LEVEL1_%TESTED", "LEVEL2_%TESTED", "LEVEL3_%TESTED", "LEVEL4_%TESTED"]]
performance_levels = performance_levels.apply(pd.to_numeric, errors = 'coerce')
level_distribution = performance_levels.mean()
print("Performance Levels Distribution: \n", level_distribution)
```

#### Math vs ELA Proficiency Rates

```python
# Ensure the proficiency percentages are numeric
data_math["PER_PROF"] = pd.to_numeric(data_math["PER_PROF"], errors="coerce")
data_ela["PER_PROF"] = pd.to_numeric(data_ela["PER_PROF"], errors="coerce")

# Calculate the mean proficiency rates
math_proficiency = data_math["PER_PROF"].mean()
ela_proficiency = data_ela["PER_PROF"].mean()

# Print the results
print(f"Math Proficiency: {math_proficiency:.2f}%")
print(f"ELA Proficiency: {ela_proficiency:.2f}%")
```

#### EM vs HS Mean Scores

```python
# Filter for EM and HS data
em_data = all_subjects[all_subjects["ASSESSMENT_NAME"].str.contains("EM", na=False)]
hs_data = all_subjects[~all_subjects["ASSESSMENT_NAME"].str.contains("EM", na=False)]
```

```python
# Compute mean scores
em_data["MEAN_SCORE"] = pd.to_numeric(em_data["MEAN_SCORE"], errors = 'coerce')
hs_data["MEAN_SCORE"] = pd.to_numeric(hs_data["MEAN_SCORE"], errors='coerce')
```

```python
# Compute mean scores
em_mean_score = em_data["MEAN_SCORE"].mean()
hs_mean_score = hs_data["MEAN_SCORE"].mean()
```

```python
print(f"EM Mean Score: {em_mean_score:.2f}")
print(f"HS Mean Score: {hs_mean_score:.2f}")
```
