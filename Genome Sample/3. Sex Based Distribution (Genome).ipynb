#### Explanation:
1. value_counts(): This counts how many male, female, or other sex categories are present in the Sex column.
2. plot(kind='bar'): A bar chart is created to visualize the counts of each sex category.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = r"E:\Data Analytics\Genome Sample\cleaned_genome_sample.xlsx"
df = pd.read_excel(file_path, header=1)
```

```python
df.head()
```

```python
# Occurence of each sex
sex_distribution = df['Sex'].value_counts()
sex_distribution
```

```python
# Visualise the sex distribution counts
plt.figure(figsize=(8,6))
sex_distribution.plot(kind='bar', color='lightcoral')
plt.title('Sex Based Distribution')
plt.xlabel('Sex')
plt.ylabel('Number of Samples')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Pie chart
plt.figure(figsize=(8, 8))
sex_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral'])
plt.title('Sex-Based Distribution (Pie Chart)')
plt.ylabel('')  # Hide the y-label for pie chart
plt.tight_layout()
plt.show()
```
