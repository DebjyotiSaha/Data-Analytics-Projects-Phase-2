#### Key Steps in the Code:
1. df['Population name'].value_counts(): This counts the number of occurrences for each unique population name in the dataset.
2. plot(kind='bar'): This generates a bar chart for the population distribution.
3. plot(kind='pie'): Optionally, this creates a pie chart to represent the distribution proportionally.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = r"E:\Data Analytics\Genome Sample\cleaned_genome_sample.xlsx"
df = pd.read_excel(file_path, header=1)
```

```python
df.head()
```

```python
# Group by 'Population name' and count the occurences
population = df['Population name'].value_counts()
population_counts = population.head(10)
population_counts
```

```python
plt.figure(figsize=(10,6))
population_counts.plot(kind='bar', color='skyblue')
plt.title('Population Distribution Analysis')
plt.xlabel('Population Name')
plt.ylabel('Number of Samples')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
population_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Population Distribution Analysis (Pie Chart)')
plt.ylabel('')  # Hide the y-label as it's not needed for the pie chart
plt.tight_layout()
plt.show()
```
