#### Analyze the sex ratio (the ratio of male to female samples) across different superpopulations.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = r"E:\Data Analytics\Genome Sample\cleaned_genome_sample.xlsx"
df = pd.read_excel(file_path, header=1)
```

```python
df.head()
```

```python
# Group by Superpopulation and Sex, then count the number of samples
sex_ratio_superpop = df.groupby(['Superpopulation name', 'Sex']).size().unstack()
```

```python
# Display the sex ratio
print(sex_ratio_superpop)
```

```python
# Visualize with a stacked bar chart
sex_ratio_superpop.plot(kind='bar', stacked=True, figsize=(12, 6), color=['lightblue', 'lightcoral'])
plt.title('Sex Ratio Across Superpopulations')
plt.xlabel('Superpopulation Name')
plt.ylabel('Number of Samples')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
