#### Compare the distribution of populations across different superpopulations (e.g., how many populations belong to each superpopulation).

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = r"E:\Data Analytics\Genome Sample\cleaned_genome_sample.xlsx"
df = pd.read_excel(file_path, header=1)
```

```python
df.head()
```

```python
# Group by Superpopulation and count populations
superpopulation_counts = df.groupby('Superpopulation name')['Population name'].nunique()
```

```python
# Display the counts of populations in each superpopulation
print(superpopulation_counts)
```

```python
# Visualize the comparison
superpopulation_counts.plot(kind='bar', figsize=(12, 6), color='lightgreen')
plt.title('Comparison of Populations Across Superpopulations')
plt.xlabel('Superpopulation Name')
plt.ylabel('Number of Populations')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
